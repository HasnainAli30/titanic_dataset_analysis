**Loading the Titanic Dataset:**

```python
import pandas as pd

titanic = pd.read_csv('/kaggle/input/titanicdatasetforanalysis/train.csv')
```

**Printing the Column Names:**

```python
titanic.columns
```

**Printing the First Five Entries:**

```python
titanic.head()
```

**Printing the number of survived men and women by using groupby() method:**

```python
survival = titanic.groupby('Sex')['Survived'].sum()   # Sum will add all the 1s as survived
survival
```

```python
# Plotting the number of men and women as survivals
import matplotlib.pyplot as plt

plt.title('Survived Men vs Survived Women')
plt.ylabel('Number of Survival')
plt.bar(['Men','Women'],[survival['male'],survival['female']], color='r', width=0.8)
```

**Counting the number of passengers who were in the 1, 2 and 3 passenger classes respectively by using groupby() function:**

```python
Pclass = titanic.groupby('Pclass')['Sex'].count()   # count will count 'male' and 'female'
Pclass
```

```python
# Plotting the number of Pclass (1,2 3) passengers

import matplotlib.pyplot as plt

plt.title('Number of Passengers Who Were in all Three Passenger Classes')
plt.xlabel('Passenger Classes')
plt.ylabel('Number of Passengers in Individual Pclass')
plt.bar(['1','2','3'],[Pclass[1],Pclass[2],Pclass[3]], color='b', width=0.75)
```

**By Using loc and value_counts() Methods to Print the Number of Male and Female Who Could Survive and Could not Survive:**

```python
titanic_male = titanic.loc[titanic['Sex'] == 'male',:]
titanic_male.head()
```

```python
titanic_male_survival = titanic_male['Survived'].value_counts()
titanic_male_survival
```

```python
titanic_female = titanic.loc[titanic['Sex'] == 'female',:]
titanic_female.head()
```

```python
titanic_female_survival = titanic_female['Survived'].value_counts()
titanic_female_survival
```

```python
# Plotting the number of survived and not survived men and women by using subplots

import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,2)

axes[0].set_title('Survival vs Not Survival of Men')
axes[0].bar(['Survived','Not Survived'],[titanic_male_survival[1], titanic_male_survival[0]], color='y')

axes[1].set_title('Survival vs Not Survival of Women')
axes[1].bar(['Survived','Not Survived'],[titanic_female_survival[1], titanic_female_survival[0]], color='g')
plt.tight_layout()  # For getting the proper spacing
```
